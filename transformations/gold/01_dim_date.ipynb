# 01 - Dim date (Bonus 5)

Builds the date dimension covering the full range of event dates in the dataset.

The range comes from min(start_date) and max(end_date) in silver - currently
1807-10-12 to 2023-01-07, giving around 79k rows. One row per calendar day,
enriched with year/month/day/weekday/quarter/weekend attributes.

*Some parts and solutions of this code was found through help from Claude*

```python
import sys
import os

# Make repo root importable so we can use utils/
sys.path.append(os.path.abspath("../.."))

from pyspark.sql import functions as F

from utils.constants import SILVER_RESULTS_OBT, GOLD_DIM_DATE
from utils.io_helpers import read_table, write_df_to_table
from utils.date_parsing import build_dim_date
```

## Find the date range and build the dimension

We look at silver to get the min start_date and max end_date,
then generate dim_date covering that range.

```python
silver_df = read_table(spark, SILVER_RESULTS_OBT)

# Get the date range from silver
date_range = silver_df.agg(
    F.min("start_date").alias("min_date"),
    F.max("end_date").alias("max_date")
).first()

min_date = date_range["min_date"]
max_date = date_range["max_date"]

print(f"Building dim_date from {min_date} to {max_date}")

dim_date_df = build_dim_date(spark, min_date, max_date)

print(f"Rows in dim_date: {dim_date_df.count():,}")
dim_date_df.show(5, truncate=False)

write_df_to_table(dim_date_df, GOLD_DIM_DATE)
print(f"Wrote {GOLD_DIM_DATE}")
```
